```python
# chose not to use KNNImpute from the sklearn.impute library because lat/lon coordinates are not linear
# euclidean distance between raw lat/lon values doesn't reflect true geographic proximity
```

```python
from sklearn.neighbors import BallTree
```

```python
def impute_by_geo(df, k=5, method='median', lat_col = "latitude",
                            lon_col="longitude", y_col="bathrooms"):
  """
  Impute NaN values in y_col using kNN based on geo distances (lat, long) with haversine metric.
  method: 'median' or 'mean' of neighbors' y values.
  """
  # Work on the original df without chained assignment
  # Ensure we have required columns
  required = [lat_col, lon_col, y_col]
  missing_cols = [c for c in required if c not in df.columns]
  if missing_cols:
    raise ValueError(f"Missing required columns: {missing_cols}")

  # Masks
  valid_geo = ((df[lat_col].notna()) & (df[lon_col].notna()))
  known_y = ((df[y_col].notna()) & (valid_geo))
  need_impute = df[y_col].isna() & valid_geo

  if known_y.sum() == 0 or need_impute.sum() == 0:
    return df # no instances to impute

  # Build BallTree on known points (converge degrees -> radians for haversine)
  coords_known_rad = np.deg2rad(
      df.loc[known_y, [lat_col, lon_col]].to_numpy()
  )
  tree = BallTree(coords_known_rad, metric="haversine")

  # Prepare targets to impute
  coords_missing_rad = np.deg2rad(
      df.loc[need_impute, [lat_col, lon_col]].to_numpy()
  )
  # Query k neighors among KNOWN points only
  # NOTE: BallTree.haversine distances are in radians (earth radius factor cancels for NN choice)
  k_eff = min(k, coords_known_rad.shape[0])
  dist_rad, idx_in_known = tree.query(coords_missing_rad, k=k_eff)

  # Pull neighbor y values and aggregate
  known_values = df.loc[known_y, y_col].to_numpy()
  if method == 'median':
    imputed_vals = np.nanmedian(known_values[idx_in_known], axis=1)
  elif method == 'mean':
    imputed_vals = np.nanmean(known_values[idx_in_known], axis=1)
  else:
    raise ValueError("method must be 'median' or 'mean'")

  # Assign back (no SettingWithCopyWarning)
  df.loc[need_impute, y_col] = imputed_vals

  return df
```

## Dropping instances in which the listed PRICE feature is an outlier

## Resorting the columns so that PRICE is appended at the end (last column)

```python
data = pd.concat([data.drop(columns=['price']), data.price], axis=1)
```

## Making a column of Z Scores for PRICE and appending at end of dataframe. Also, making a boolean column that states whether the listed price is an outlier, depending on whether its PRICE lies outside of +/-3 Standard Deviations from the MEAN price

```python
data = data.assign(
    price_ZScore = lambda x: (x.price.sub(x.price.mean())).div(x.price.std()),
    outlier = lambda x: np.where(abs(x.price_ZScore ) > 3, 1, 0)
)
```

```python
data.shape[0] # rows before dropping outliers
```

```python
drop_rows = data.query('outlier==1').index # obtaining indexes where the instance has an OUTLIER PRICE LISTING
```

```python
drop_rows.shape[0] # number of instances in LISTING dataset that has OUTLIER PRICE LISTING
```

```python
data.drop(drop_rows, axis=0, inplace=True) # dropping outlier instances
```

```python
data.drop(['price_ZScore', 'outlier'], axis=1, inplace = True) # dropping derived columns
```
