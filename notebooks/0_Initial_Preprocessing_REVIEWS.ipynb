```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('/content/reviews.csv.gz')
```

```python
data.info()
```

```python
data.head()
```

```python
data = data.assign(
    listing_id = lambda x: x.listing_id.astype("object"),
    id = lambda x: x.id.astype("object"),
    date = lambda x: pd.to_datetime(x.date),
    reviewer_id = lambda x: x.reviewer_id.astype("object")
)
```

```python
data.info()
```

```python
data.to_csv('cleaned_reviews', index=False)
```
